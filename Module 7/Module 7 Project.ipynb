## NYPD Motor Vehicle Collision Data

### Overview

The Motor Vehicle Collisions Crash table contains details on the crash events. Each row represents a crash event. The Motor Vehicle Collisions data tables contain information from all police reported motor vehicle collisions in NYC. The dataset can be found by following this link: https://data.cityofnewyork.us/Public-Safety/NYPD-Motor-Vehicle-Collisions-Crashes/h9gi-nx95

### High-Level Description

The data dates from 2012 to the current day, with data being updated on a daily basis. At the time of this writing, there are 1.59 million rows, each row representing a crash event, and over 29 columns which represent date, time, borough, zip code, latitude, longitude, location, on and off street name, cross street name, number of persons injured, number of persons killed, number of pedestrians injured, number of pedestrians killed, number of cyclist injured, number of cyclist killed, number of motorist injured, number of motorist killed, contributing factors, vehicle type codes and collision ID.

### Bring in the data

Okay, now, before starting my data visulization, I will transform the string timestamp for 'date' to a true datetime data type.

```python
datanyc['date'] = pd.to_datetime(datanyc['date'])
datanyc.head()
```

Next, I will change the the 'persons_injured' and 'persons_killed' columns from float to integer. To do so, I will first get rid of the missing values in the 'persons_injured' and 'persons_killed' columns.

```python
datanyc.dropna(subset = ['persons_injured'], how='all', inplace=True)
datanyc.dropna(subset = ['persons_killed'], how='all', inplace=True)
```

```python
datanyc['persons_injured'] = datanyc.persons_injured.astype(int)
datanyc['persons_killed'] = datanyc.persons_injured.astype(int)
datanyc.head()
```

## Visualization With Matplotlib

Let's see the average of killed persons by borough with the collisions count.

```python
datanyc.groupby('borough').persons_killed.agg(['count', 'mean'])
```

Let's graph this with a bar chart.

```python
import matplotlib.pyplot as plt
%matplotlib inline
datanyc.groupby('borough').persons_killed.mean().plot(kind='bar', figsize=(10, 6))
plt.ylabel('average persons killed')
plt.title('Average of Deaths by Borough', fontsize=15)
plt.xticks(rotation=0, fontsize='large')
plt.xlabel(' ')
```

It looks like Brooklyn and Bronx have the highest number of deaths (nearly the same) meanwhile Manhattan has the lowest number of deaths, which can be due to its smaller size.

## Visualization With Seaborn

I want to see whether there is a relationship between hour and persons killed. To do so, I will create a new column called 'hour' in which I will only have the hours instead of hours and minutes. 

```python
datanyc['time'] = pd.to_datetime(datanyc.time)
datanyc['hour'] = datanyc['time'].dt.hour
datanyc.head()
```

```python
import seaborn as sns
sns.set(style="whitegrid")

f, ax = plt.subplots(figsize=(20, 10))
sns.scatterplot(x="persons_killed", y="hour",
                linewidth=0, s=150,
                data=datanyc, ax=ax)
```

In most collisions 0 to 5 people die and generally the collisions in which more than 15 people die occur between around 7:00 and 18:00. Interesting.

## Visualization With Pandas

Let's see the mean of persons killed by hour. I am interested in seeing the percentage rate of persons killed for each hour.

```python
datanyc.groupby('time').persons_killed.mean().plot(figsize=(10, 6))
plt.title('Average Number of Deaths by Hour', fontsize=18)
plt.ylabel('average number of persons killed', fontsize=13)
plt.xlabel('time', fontsize=13)
```

Hmm.. I rather see this information hourly to get a better general idea. To do so, I will first drop the missing values in the 'date' and 'time' columns and change their data types from integer to string. Then, create a new column named 'combined', which will combine 'date' and 'time' columns into the same panda series. Finally, take that series and convert it to datetime data type and store it in a new column called 'combined_date_time'. That will allow me to get the hours when graphing it.

```python
datanyc.dropna(subset = ['date'], how='all', inplace=True)
datanyc.dropna(subset = ['time'], how='all', inplace=True)
datanyc['date'] = datanyc.date.astype(str)
datanyc['time'] = datanyc.time.astype(str)
datanyc.head()
```

```python
combined = datanyc.date.str.cat(datanyc.time, sep= ' ')
combined.head()
```

```python
datanyc['combined_date_time'] = pd.to_datetime(combined)
```

```python
datanyc.dtypes
```

```python
datanyc.groupby(datanyc.combined_date_time.dt.hour).persons_killed.mean().plot(figsize=(10, 6))
plt.title('Average Number of Deaths by Hour', fontsize=18)
plt.ylabel('average number of persons killed', fontsize=13)
plt.xlabel('time', fontsize=13)
```

It looks like number of deaths goes up around 4:00 in the morning and 23:00 in the evening. It decreases during the day between 8:00 and 14:00, which are typical work hours.
