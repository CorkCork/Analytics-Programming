In this notebook, I am ingesting a fixed width file that deals with precipitation measurements. This data is used to spot trends in precipitation patterns.

This dataset is provided by the 'National Oceanic and Atmospheric Administration, Department of Commerce'. It includes the monthly U.S. reported precipitation amounts in hundredths of inches (ex 100 is 1.00 inches) generated from the GTS metar(hourly) and synoptic(6-hourly) observations for selected cities based on the Weekly Weather and Crop Bulletin station list. More information can be found by following this link: https://catalog.data.gov/dataset/climate-prediction-center-cpc-monthly-u-s-selected-cities-precipitation-summary

```python
import pandas as pd
```

I already opened the file in a text editor and discovered that the first __ rows are header/descriptive rows that I want to ignore and not read in.

```python
precip_df = pd.read_fwf('C:/Users/Inspiration/Desktop/module6/mctyprcp.txt', skiprows=10, header=None,
                        colspecs='infer', widths=None)
precip_df.head()
```

And let's look at the end of our data!

```python
precip_df.tail()
```

The last row of data (249) is the "end of data" and shouldn't be included. Let's get rid of it!

```python
precip_df = precip_df.drop(249)
precip_df.tail()
```

Also, the second column ("1") has two values in it. We can split this column easily because we know that the last character (position -1) is the "pflag" and that everything else belongs to the city name. Let's create two new columns, city and pflag, which are the result of splitting the combined column.

```python
precip_df['city'] = precip_df[1].str[:-1]  
precip_df['pflag'] = precip_df[1].str[-1] 
```

Let's check it!

```python
precip_df.head()
```

Let's remove column 1, since we've captured its data elsewhere.

```python
precip_df = precip_df.drop(columns=[1])
precip_df.head()
```

It looks like column 10 also needs to be split! We can do this again by position, first 4 (up to but not including position 5) and last 4 (starting with position 5 and continuing until the end)

```python
precip_df['year_to_date_amt'] = precip_df[10].str[:5] 
precip_df['year_to_date_dep_nml'] = precip_df[10].str[5:] 

precip_df = precip_df.drop(columns=[10])
precip_df.head()
```

Now, all we have to do are supply column names and maybe re-order things so that the city and pflag come as the second and third columns, respectively:

```python
precip_df.rename(columns={0:'state',
                          2: "call_ltrs",
                          3: "month_tot_amt",
                          4: "month_dep_nml",
                          5: "24_hr_max",
                          6: "days_gt_1",
                          7: "days_gt_50",
                          8: "season_to_date",
                          9: "season_dep_nml",
                          11: "year_to_date_pct_nml"},inplace=True)

precip_df = precip_df[["state",
                       "city",
                       "pflag",
                       "call_ltrs",
                       "month_tot_amt",
                       "month_dep_nml",
                       "24_hr_max",
                       "days_gt_1",
                       "days_gt_50",
                       "season_to_date",
                       "season_dep_nml",
                       "year_to_date_amt",
                       "year_to_date_dep_nml",
                       "year_to_date_pct_nml"]]
```

```python
precip_df.head()
```

Now let's identify the city with the highest and lowest year_to_date_pct_nml. We'll sort first, then we can just look at the first and the last row.

```python
precip_df = precip_df.sort_values(by='year_to_date_pct_nml', ascending=True)
```
